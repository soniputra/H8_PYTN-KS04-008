# Area Plot

## Import Library

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Plot : Area

```python
df_tahun_2013 = df_not_null_data[df_not_null_data["year"].isin([2013])]
df_tahun_2013
```

```python
df_maj_year = df_tahun_2013.groupby(["major_category", "borough"]).sum()["value"].sort_values(ascending=False) 
```

```python
df_maj_year.unstack().T #transpose
```

```python
fig, ax = plt.subplots(figsize=(20,10))
df_maj_year.unstack().T.plot(kind="area",ax=ax,stacked=False)

plt.title('total') # add a title to the area plot
plt.ylabel('value') # add y-label
plt.xlabel('year') # add x-label

plt.legend(loc='upper left')

plt.show()
```

##### Keterangan

Dari Grafik Tipe Area, dapat dilihat bahwasannya tingkat kejahatan yang terjadi di setiap wilayah (borough) pada tahun 2013 adalah Other Notifiable Offences

## Plot : Bar 

```python
# Stage dataframe for plotting

df_area = df_tahun_2013.groupby(['borough']).sum()['value'].sort_values(ascending=False)
df_area.plot(kind='bar', figsize=(20,8))

# Set axis to the graphs
plt.title("Borough crime in the 2013") # add a title to the area plot
plt.ylabel('Total Kejahatan') # add y-label
plt.xlabel('Borough') # add x-label

plt.show()
```

##### Keterangan

Dari Grafik Tipe Bar, dapat dilihat bahwasannya tingkat kejahatan yang terjadi pada tahun 2013 sering terjadi di wilayah (borough) Croydon dan kejahatan yang jarang terjadi pada tahun 2013 berada di wilayah City of London

```python
df_maj = df_tahun_2013.groupby(['month']).sum()['value'].sort_values(ascending=True)

df_maj.plot(kind="bar", figsize=(10,8))
plt.xticks(rotation=40)
plt.title("Kejahatan Pada Tahun 2013 Berdasarkan Bulan") # add a title to the area plot
plt.ylabel("Tingkat Kejahatan") # add y-label
plt.xlabel("Bulan") # add x-label

plt.show()
```

##### Keterangan

Dan untuk kejahatan yang sering terjadi berada di bulan ke 7 dan tindak kejahatan terendah berada di bulan ke 2 pada tahun 2013

## Plot : Pie

```python
# Stage dataframe for plotting

df_area = df_tahun_2013.groupby(['major_category']).sum()['value'].sort_values(ascending=False)
df_area.plot(kind='pie', figsize=(20,8), autopct='%1.0f%%')

# Set axis to the graphs
plt.title("Borough crime in the 2013") # add a title to the area plot
plt.ylabel('Total Kejahatan') # add y-label
plt.xlabel('Borough') # add x-label

plt.show()
```

# Word Cloud

## Import Library

```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
df_wordclouds = df[df['year'].isin([2013])]
# df_woc = df_woc[df_woc['borough'].isin(['City of London'])]
df_wordclouds = df_wordclouds['major_category'].str.cat(sep=' ')

df_wordcloud = WordCloud(
    background_color='white',
    stopwords=stopwords
)

# generate word cloud
df_wordcloud.generate(df_wordclouds)

fig = plt.figure()
# set width
fig.set_figwidth(14)
# set height
fig.set_figheight(18)

plt.imshow(df_wordcloud, interpolation='bilinear')

plt.axis('off')
plt.show()
```
